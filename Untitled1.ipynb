```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
import seaborn as sns
from sklearn import  svm
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('HR-Employee-Attrition.csv',encoding="latin1")
df.head()
#df=pd.read_csv('C:\\Users\\Mr.Bilal\\Desktop\\HR-Employee-Attrition.csv',usecols=["Age"])
#df.sort_index()
Gender=df[['Gender']]
df=df.replace({'Gender':'Female','Attrition':'Yes'},0)
df=df.replace({'Gender':'Male','Attrition':'No'},1)
Business=df[['BusinessTravel']]
df=df.replace({'BusinessTravel':'Travel_Rarely'},0)
df=df.replace({'BusinessTravel':'Travel_Frequently'},1)
df=df.replace({'BusinessTravel':'Non-Travel'},2)
              

df=df.replace({'Department':'Sales'},0)
df=df.replace({'Department':'Research & Development'},1)
df=df.replace({'Department':'Human Resources'},2)  

df=df.replace({'EducationField':'Life Sciences'},0)
df=df.replace({'EducationField':'Medical'},1)
df=df.replace({'EducationField':'Other'},2)
df=df.replace({'EducationField':'Technical Degree'},3)
df=df.replace({'EducationField':'Marketing'},4)
df=df.replace({'EducationField':'Human Resources'},5)

df=df.replace({'JobRole':'Sales Representative'},0)
df=df.replace({'JobRole':'Sales Executive'},1)
df=df.replace({'JobRole':'Research Scientist'},2)
df=df.replace({'JobRole':'Research Director'},3)
df=df.replace({'JobRole':'Manufacturing Director'},4)
df=df.replace({'JobRole':'Manager'},5)
df=df.replace({'JobRole':'Laboratory Technician'},6)
df=df.replace({'JobRole':'Human Resources'},7)
df=df.replace({'JobRole':'Healthcare Representative'},8)

df=df.replace({'MaritalStatus':'Single'},0)
df=df.replace({'MaritalStatus':'Married'},1)
df=df.replace({'MaritalStatus':'Divorced'},2)

df=df.replace({'Over18':'Y'},0)

df=df.replace({'OverTime':'Yes'},0)
df=df.replace({'OverTime':'No'},1)
print(df)
y=df["Attrition"]
x=df.drop("Attrition",1)
x=x.drop("Age",1)
x=x.drop("EmployeeCount",1)
x=x.drop("EmployeeNumber",1)
x=x.drop("Over18",1)


#df=df[df["Attrition"]]


```

```python
rf=RandomForestClassifier(n_estimators=500,random_state=123)

x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=123)


rf.fit(x_train,y_train)	
acc=rf.score(x_test,y_test)
print(acc)

pred=rf.predict(x_test)
print(pred)
cm=confusion_matrix(y_test,pred) 

TP =cm[0][0]
TN= cm[0][1]
FP=cm[1][0]
FN=cm[1][1]
print(TP)
print(TN)
print(FP)
print(FN)
plt.figure(figsize=(8, 4))
plt.clf()
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.cool)
classNames = ['yes','no']
plt.title('Random \n Accuracy:{0:.3f}'.format(acc))
plt.ylabel('True label')
plt.xlabel('Predicted label')
tick_marks = np.arange(len(classNames))
plt.xticks(tick_marks, classNames, rotation=45)
plt.yticks(tick_marks, classNames)
s = [['TN','FP'], ['FN', 'TP']]
for i in range(2):
    for j in range(2):
        plt.text(j,i, str(s[i][j])+" = "+str(cm[i][j]))
plt.show()
print(df['Attrition'].corr(df['OverTime']))
df['Attrition'].corr(df['DistanceFromHome'])
df['Attrition'].corr(df['RelationshipSatisfaction'])
df['Attrition'].corr(df['WorkLifeBalance'])
df['Attrition'].corr(df['PerformanceRating'])
df['Attrition'].corr(df['TotalWorkingYears'])
df['Attrition'].corr(df['YearsWithCurrManager'])
df['Attrition'].corr(df['PerformanceRating'])
df['Attrition'].corr(df['DailyRate'])
df['Attrition'].corr(df['EnvironmentSatisfaction'])
df['Attrition'].corr(df['JobInvolvement'])
df['Attrition'].corr(df['TotalWorkingYears'])
```
